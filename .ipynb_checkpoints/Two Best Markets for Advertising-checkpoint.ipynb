## Introduction
In this project, we assume that we are working for an e-learning company that focuses on programming courses. Our goal is to provide the marketing team with insight into which countries would have the most appetite for our products. We need to find the two best markets to advertise our product based on the size of the markets and the average amount of money the coders are willing to spend.

The best way to gather relevant data would be to conduct surveys in each market. However, since this would be an expensive and slow process, we decided to explore already available datasets online. In particular, we decided to use the data from [freeCodeCamp's 2017 New Coder Survey](https://github.com/freeCodeCamp/2018-new-coder-survey). This dataset contains responses from more than 20,000 developers in multiple markets that we are interested in. 


### Read in Data

```python
import pandas as pd
import numpy as np
from IPython.display import display

survey = pd.read_csv('C:/Users/Chris Han/2017-new-coder-survey/clean-data/2017-fCC-New-Coders-Survey-Data.csv', low_memory = False)
print(survey.shape)
survey.head()
pd.options.display.max_columns = None
display(survey)
```

The columns of most interest to us are the job interest columns and MoneyForLearning column. We are using the clean data instead of the raw data for convenience. The data set combines information from two surveys given out to coders to gain insight onto what medium of education they used to achieve their goals and how much money they spent on it. 

### Data Exploration

Our population of interest is new coders who may be interested in the programs that we offer. In particular, we want to check where these new coders are located and how much money they are willing to spend on learning. Before we begin, we need to look at whether the sample we have is representative of the population of interest. 

```python
jobinterest_freq = survey['JobRoleInterest'].value_counts(normalize = True) * 100

import matplotlib.pyplot as plt
%matplotlib inline

jobinterest_freq
```

It seems that most people are interested in one subject. Let's explore how the number of subjects that people wish to learn is distributed.

```python
# Split each string in the 'JobRoleInterest' column
interests_no_nulls = survey['JobRoleInterest'].dropna()
splitted_interests = interests_no_nulls.str.split(',')

# Frequency table for the var describing the number of options
n_of_options = splitted_interests.apply(lambda x: len(x)) # x is a list of job options
n_of_options.value_counts(normalize = True).sort_index() * 100
```

As we can see, 31.65% of the people are focused on one subject they want to learn, whereas the rest are interested in two or more subjects. This is not a bad sign after all since new coders should be interested in exploring multiple topics. Because most of our products are focused on web or mobile development, let's find out the percentage of people interested in those two.

```python
# Frequency table
web_or_mobile = interests_no_nulls.str.contains('Web Developer|Mobile Developer') # returns an array of booleans

web_or_mobile.value_counts(normalize = True) * 100

plt.style.use('fivethirtyeight')
(web_or_mobile.value_counts(normalize = True) * 100).plot.bar(title = "Percentage of people interested in web or mobile development"
                                                             )
```

The percentage of people in our sample that are interested in either web or mobile development courses is about 86%. Again, our goal is to target our advertisements to those who are interested in web or mobile development since the majority of our products are focused on those two subjects. Based on the results, it seems that the sample we have is an appropriate representation of the population of interest, and we can continue with our analysis.

## Which country is most appropriate for our advertisements?
Now that we know that we have a good dataset to work with, we move on to the question of which market will serve us the best in advertising our courses. We want to find the markets with the largest number of new coders.

### Drop missing values

Before we get started, we will remove all the rows with missing value for the Job Role Interest column as they are needed for our analysis.

```python
survey_good = survey[survey['JobRoleInterest'].notnull()].copy()
```

```python
country_counts = survey_good['CountryLive'].value_counts()
country_prop = survey_good['CountryLive'].value_counts(normalize = True)
```

```python
print(country_counts.head(4), '\n\n', country_prop.head(4))
```

United States of America has the largest percentage of people that answered the survey with 45%, and India comes in second at 8%. Based on this, we get a sense that U.S and India may be the two markets to advertise in. However, we still need to look at whether the people in the two markets are interested in web and mobile development courses and how much money they are willing to spend.

It also seems like a good idea to narrow down our analysis to only four countries: the US, India, the United Kingdom, and Canada. Two reasons for this decision are:

- These are the countries having the highest absolute frequencies in our sample, which means we have a decent amount of data for each.
- Our courses are written in English, and English is an official language in all these four countries. The more people that know English, the better our chances to target the right people with our advertisements.


## How much money are coders willing to spend?


```python
# Replace 0 with 1 since 0 month programming usually means they just started
survey_good['MonthsProgramming'].replace(0, 1, inplace = True)

# Find the money spent per month
survey_good['MoneySpentMonthly'] = survey_good['MoneyForLearning'] / survey_good['MonthsProgramming']

# Check for null values
sum(survey_good['MoneySpentMonthly'].isnull())

# Remove the rows with null values
survey_good = survey_good[survey_good['MoneySpentMonthly'].notnull()]

# Also remove the rows with null values for the Country Live column
survey_good = survey_good[survey_good['CountryLive'].notnull()]

# Find the average monthly spending for coding classes for students in U.S, India, U.K, and Canada

countries_of_interest = ['United States of America',
                         'India',
                         'United Kingdom',
                         'Canada']
countries_means = survey_good.groupby('CountryLive').mean()
four_means = countries_means['MoneySpentMonthly'][countries_of_interest]

# Compare with Median

countries_medians = survey_good.groupby('CountryLive').median()
four_medians = countries_medians['MoneySpentMonthly'][countries_of_interest]

print(four_means)
print(four_medians)
print(survey_good['CountryLive'].value_counts())
```

The results for the United Kingdom and Canada are a bit surprising relative to the values we see for India. If we considered a few socio-economical metrics (like GDP per capita), we would intuitively expect people in the UK and Canada to spend more on learning than people in India.

There are few possibilities:
- we don't have have enough representative data for the United Kingdom and Canada
- we have some outliers (maybe coming from wrong survey answers) making the mean too large for India, or too low for the UK and Canada. 
- the results are correct as is

Next, we will remove outliers to see how it affects our results.

## Remove outliers

```python
# isolate just the countries of interest
four_countries = survey_good[survey_good['CountryLive'].str.contains('United States of America|India|United Kingdom|Canada')]

import seaborn as sns

sns.boxplot(y = 'MoneySpentMonthly',
           x = 'CountryLive',
           data = four_countries)

plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

It looks like there are three outliers, two in U.S and one in India, that we may want to remove. First let's remove the most obvious outliers in U.S.

```python
four_countries = four_countries[four_countries['MoneySpentMonthly'] < 20000]
```

```python
sns.boxplot(y = 'MoneySpentMonthly',
           x = 'CountryLive',
           data = four_countries)

plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

Let's examine the outliers for India.

### India outliers

```python
india_outliers = four_countries[(four_countries['CountryLive'] == 'India') &
                  (four_countries['MoneySpentMonthly'] > 2500)]

four_countries[(four_countries['CountryLive'] == 'India') &
              (four_countries['MoneySpentMonthly'] > 2500)]
```

Interestingly, the respondents who had the highest money spent per month did not attend bootcamps which tend to be the most expensive. We suspect that this was an user error and we will remove these rows.

```python
four_countries = four_countries.drop(india_outliers.index)
```

### U.S Outliers

```python
us_outliers = four_countries[(four_countries['CountryLive'] == 'United States of America') &
                  (four_countries['MoneySpentMonthly'] > 6000)]

four_countries[(four_countries['CountryLive'] == 'United States of America') &
              (four_countries['MoneySpentMonthly'] > 6000)]
```

In the next code block, we will remove respondents that:

- Didn't attend bootcamps.
- Had been programming for three months or less when at the time they completed the survey.


```python
no_bootcamp = four_countries[(four_countries['CountryLive'] == 'United States of America') &
                            (four_countries['AttendedBootcamp'] == 0) &
                            (four_countries['MoneySpentMonthly'] > 6000)]

four_countries = four_countries.drop(no_bootcamp.index)

three_months_less = four_countries[(four_countries['CountryLive'] == 'United States of America') &
                                    (four_countries['MoneySpentMonthly'] > 6000) &
                                      (four_countries['MonthsProgramming'] <= 3)]
four_countries = four_countries.drop(three_months_less.index)
```

```python
sns.boxplot(y = 'MoneySpentMonthly',
           x = 'CountryLive',
           data = four_countries)

plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

## Canada Outlier
Now we are able to see that there is also an outlier for Canada. Let's examine it.

```python
canada_outlier = four_countries[(four_countries['CountryLive'] == 'Canada') &
                               (four_countries['MoneySpentMonthly'] > 3000)]
four_countries[(four_countries['CountryLive'] == 'Canada') &
                               (four_countries['MoneySpentMonthly'] > 3000)]
```

Similary to the respondents from U.S, this person only took 2 months of programming and spent $10000, signaling that the individual spent a large sum of money upfront for the coding camp but will not spend more later, hence erroneously inflating the money spent per month. Let's remove it.

```python
four_countries = four_countries.drop(canada_outlier.index)
```

```python
sns.boxplot(y = 'MoneySpentMonthly',
           x = 'CountryLive',
           data = four_countries)

plt.title('Money Spent Per Month Per Country\n(Distributions)',
         fontsize = 16)
plt.ylabel('Money per month (US dollars)')
plt.xlabel('Country')
plt.xticks(range(4), ['US', 'UK', 'India', 'Canada']) # avoids tick labels overlap
plt.show()
```

### Average money spent monthly and proportion of new coders in each market after removing outliers

```python
# Mean
countries_means2 = four_countries.groupby('CountryLive').mean()

print(countries_means2['MoneySpentMonthly'])
```

We see a slight change here in that now Canada has the second highest average money spent monthly. To finish our analysis, let's take another look at the proportion of new coders in each country and see how it changed after removing the outliers.

```python
print(four_countries['CountryLive'].value_counts(normalize = True))
```

## Choosing the two best markets to advertise

We have seen that U.S has the largest representation of new coders in our sample with about 75% after removing the outliers. U.S also has the higest average money spent monthly of about $143. Hence, it is an obvious choice for the best market to advertise in for our courses. For the second best market, the answer is not as apparent.

When looking at the proportion of new coders, India has the second largest market with about 12%. However, Canada has the second highest average monthly spending on learning with about $93. The question becomes whether it would be better to choose India, which has a larger market, or Canada, where people are likely to spend more on learning coding. Considering that our courses costs $59 per month, the average consumer in India would still be able to afford it since it is within their budget of around $66. 

Perhaps it is possible to market in all three countries. For example, we can use 75% of our resources in U.S, 12% in India, and 6% in Canada. Nevertheless, there is a risk of spreading ourselves too thin by tapping into all three markets in this manner. Another option is to concentrate our efforts in the two biggest markets, namely U.S and India (70%, 30%, respectively). 

The best decision would be to send our results to the marketing team and leave it upto their expertise.
